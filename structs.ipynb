```python
import numpy
import random
import timeit
```

```python
samplelst = [8, 12, 12, 54, 133, "abs", "cbd"] * 1000
class single_lst:
    def __init__(self, lst):
        self.obj = {"0": [None, "0"]}
        self.size = 0
        nxt = "0"
        for item in lst:
            i = "1"
            while i in self.obj:
                 i = str(random.random())
            self.obj[nxt][1] = i
            self.obj[i] = [item, "0"]
            nxt = i
            self.size += 1

    def value(self):
        index = "0"
        lst = []
        for i in range(self.size):
            index = self.obj[index][1]
            lst.append(self.obj[index][0])
        return lst

    def search(self, **kwargs):
        if "value" in kwargs:
            target = kwargs["value"]
            nxt = self.obj["0"][1]
            while nxt != "1":
                if target == self.obj[nxt][0]:
                    return nxt
                else:
                    nxt = self.obj[nxt][1]
            return None
        if "position" in kwargs:
            target = kwargs["position"]
            if target+1 > self.size:
                raise IndexError(f"Index out of range. Size is {self.size}, requested is {target+1}.")
            nxt = self.obj["0"][1]
            for i in range(target):
                nxt = self.obj[nxt][1]
            return nxt, self.obj[nxt][0]

    def insert(self, position, value):
        try:
            if position == 0:
                n1 = "0"
            else:
                n1,_ = self.search(position = position-1)
        except:
            raise
        i = "0"
        while i in self.obj:
                 i = str(random.random())
        self.obj[i] = [value, self.obj[n1][1]]
        self.obj[n1][1] = i
        self.size += 1

    def remove(self, position):
        try:
            if position == 0:
                n1 = "0"
            else:
                n1, _ = self.search(position = position-1)
            n2 = self.obj[n1][1]
            if n2 == "0":
                raise IndexError(f"Index out of range. Size is {self.size}, requested is {position+1}.")
        except:
            raise
        self.obj[n1][1] = self.obj[n2][1]
        self.size -= 1
        del self.obj[n2]
```

```python
class single_stack(single_lst):
    def __init__(self, lst):
        super().__init__(lst)
    
    def push(self, value):
        self.insert(0, value)

    def take(self):
        n1, value = self.search(position = 0)
        self.remove(0)
        return value
```

```python
class double_lst:
    
    def __init__(self, lst):
        self.obj = {"0": ["0", None, "0"]}
        self.size = 0
        nxt = "0"
        for item in lst:
            i = "1"
            while i in self.obj:
                 i = str(random.random())
            self.obj[nxt][2] = i
            self.obj[i] = [nxt, item, "0"]
            nxt = i
            self.size += 1
        self.obj["0"][0] = nxt

    def value(self):
        index = "0"
        lst = []
        for i in range(self.size):
            index = self.obj[index][2]
            lst.append(self.obj[index][1])
        return lst

    def search(self, **kwargs):
        if "value" in kwargs:
            target = kwargs["value"]
            nxt = self.obj["0"][2]
            while nxt != "1":
                if target == self.obj[nxt][1]:
                    return nxt
                else:
                    nxt = self.obj[nxt][2]
            return None
        if "position" in kwargs:
            target = kwargs["position"]
            if target+1 > self.size:
                raise IndexError(f"Index out of range. Size is {self.size}, requested is {target+1}.")
            if self.size//2 > target:
                direction = 1
            else:
                direction = -1
                target = self.size - target - 1
            nxt = self.obj["0"][1+direction]
            for i in range(target):
                nxt = self.obj[nxt][1+direction]
            return nxt, self.obj[nxt][1]

    def insert(self, position, value):
        try:
            if position == 0:
                n1 = "0"
            else:
                n1,_ = self.search(position = position-1)
        except:
            raise
        i = "0"
        while i in self.obj:
                 i = str(random.random())
        self.obj[i] = [n1, value, self.obj[n1][2]]
        self.obj[self.obj[n1][2]][0] = i
        self.obj[n1][2] = i
        self.size += 1

    def remove(self, position):
        try:
            n1, _ = self.search(position = position)
        except:
            raise
        n0 = self.obj[n1][0]
        n2 = self.obj[n1][2]
        self.obj[n0][2] = n2
        self.obj[n2][0] = n0
        self.size -= 1
        del self.obj[n1]
```

```python
samplelst = [8, 12, 12, 54, 133, "abs", "cbd"] * 100000

sq = single_queue(samplelst)
dq = double_queue(samplelst)
count1 = int(1e6)
count2 = int(1e2)
t1 = timeit.Timer("sq.push(14)", globals = globals())
t2 = timeit.Timer("dq.push(14)", globals = globals())

t3 = timeit.Timer("sq.take()", globals = globals())
t4 = timeit.Timer("dq.take()", globals = globals())

print(t1.timeit(count1)/count1)
print(t2.timeit(count1)/count1)

print(t3.timeit(count2)/count2)
print(t4.timeit(count1)/count1)
```
